```python
import os
import openai
import dotenv
import sys

dotenv.load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

def moderate(question):
    
    cats = {
        "hate": "Content that expresses, incites, or promotes hate based on race, gender, ethnicity, religion, nationality, sexual orientation, disability status, or caste.",
        "hate/threatening": "Hateful content that also includes violence or serious harm towards the targeted group.",
        "self-harm": "Content that promotes, encourages, or depicts acts of self-harm, such as suicide, cutting, and eating disorders.",
        "sexual": "Content meant to arouse sexual excitement, such as the description of sexual activity, or that promotes sexual services (excluding sex education and wellness).",
        "sexual/minors": "Sexual content that includes an individual who is under 18 years old.",
        "violence": "Content that promotes or glorifies violence or celebrates the suffering or humiliation of others.",
        "violence/graphic": "Violent content that depicts death, violence, or serious physical injury in extreme graphic detail.",
    }
    
    response = openai.Moderation.create(input = question)
    output = response["results"][0]
    
    if(output.flagged):
        print('\033[94m' + "\n\n***WARNING*** Your questions didn't pass the following moderation checks:- \n")
        i=0;
        for cat, text in cats.items():
            if(output.categories[cat]):
                i=i+1
                print('\033[94m' + str(i) + ". " + cat + ":" + text)
                
        print('\033[93m' + "\n_____________________________________________________________________________________________________________________________\n\n")
        return True
    
    return False
        
    
    
    

prompt_text = "You are an AI Recipe generator.\nYou generate recipes using the ingredients or the prompt provided by the user.\n\nyou can use other readily available ingredients too, but its compulsory to use the ingredients provided by the user.\n\nIf user prompts you that he/she does not have a certain ingredient, provide an alternative for it or generate another recipe excluding it.\n\nAdditionally you need to provide nutritional values in the recipe if user asks for it.\n\nAlso keep in mind for keywords like veg, non-veg or vegan.\n\nIf you don't have an answer  for the question or it doesn't fit the above purpose, strictly respond with phrase: \"I'm only a simple AI Chatbot, configured to assist you with different recipes. Sorry but  I cant help you with that!\".\n"

msg=[
        {"role":"system", "content":prompt_text},
    ]
    
print('\033[92m' + "Bot: How can I help you? ")
sys.stdout.flush()

while(True):
    
    print('\033[91m')
    question=input("User: ")
    
    if(moderate(question)):
        continue
        
    
    print('\033[93m' + "Please wait till I fetch the data.........." + "\n")
    
    msg.append({"role":"user", "content":question})

    response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      messages=msg,
      temperature=0.6,
      max_tokens=2000,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )
    

    answer=response.choices[0].message.content
    msg.append({"role":"assistant", "content":answer})

    print('\033[92m' + "Bot: "+answer)
    print('\033[93m' + "\n_____________________________________________________________________________________________________________________________\n\n")

```
